```python

import numpy as np

# Create a 4x4 array from an exponential distribution
array_4x4 = np.random.exponential(scale=1.0, size=(4, 4))
print("4x4 Exponential Random Array:\n", array_4x4)
```

```python
import matplotlib.pyplot as plt

# Generate 100000 samples from exponential, uniform, and normal distributions
exp_data = np.random.exponential(scale=1.0, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

# Plot histogram for all three distributions
plt.figure(figsize=(10, 6))

plt.hist(exp_data, density=True, bins=200, histtype="step", color="green", label="Exponential")
plt.hist(uniform_data, density=True, bins=200, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=200, histtype="step", color="red", label="Normal")

plt.axis([-2, 8, 0, 1.5])
plt.legend(loc="upper right")
plt.title("Random Distributions: Exponential, Uniform, Normal")
plt.xlabel("Value")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Define the grid range
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Compute Z = X² + Y²
Z = X**2 + Y**2

# Plot 3D surface
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title('3D Plot of Z = X² + Y²')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import spearmanr

# Load the penguins dataset from seaborn
df = sns.load_dataset("penguins").dropna()

# Select relevant features (using numerical columns from 'penguins')
columns = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
df_sub = df[columns]

# Pearson correlation
pearson_corr = df_sub.corr(method='pearson')

# Spearman correlation
spearman_corr, _ = spearmanr(df_sub)

# Convert Spearman result to DataFrame
spearman_corr_df = pd.DataFrame(spearman_corr, index=columns, columns=columns)

# Plot Pearson heatmap
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', square=True)
plt.title("Pearson Correlation")

# Plot Spearman heatmap
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr_df, annot=True, cmap='YlGnBu', square=True)
plt.title("Spearman Correlation")

plt.tight_layout()
plt.show()
```
